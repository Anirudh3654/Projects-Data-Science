```python
import string
string.punctuation
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
stopwords.words('english')
def clean_data(data):
    punc_removed=[s for s in data if s not in string.punctuation]
    punc_removed_join=" ".join(punc_removed)
    punc_removed_cleaned=[s for s in punc_removed_join if s.lower() not in stopwords.words('english')]
    return punc_removed_cleaned
```

```python
reviews_cleaned=reviews['verified_reviews'].apply(clean_data)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
vectorizer=CountVectorizer(analyzer=clean_data)
reviews_countvectorizer=vectorizer.fit_transform(reviews['verified_reviews'])
```

```python
print(vectorizer.get_feature_names())
```

```python
print(reviews_countvectorizer.toarray())
```

```python
reviews_countvectorizer.shape
```

```python
r=pd.DataFrame(reviews_countvectorizer.toarray())
```

```python
X=r
```

```python
y=reviews['feedback']
y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

```python
from sklearn.naive_bayes import MultinomialNB
NB_classifier=MultinomialNB()
NB_classifier.fit(X_train,y_train)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
y_predict_test=NB_classifier.predict(X_test)
cm=confusion_matrix(y_test,y_predict_test)
sns.heatmap(cm,annot=True)
```

```python
print(classification_report(y_test,y_predict_test))
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
model.fit(X_train,y_train)
y_pred=model.predict(X_test)
cm=confusion_matrix(y_pred,y_test)
sns.heatmap(cm,annot=True)
```

```python
from sklearn.ensemble import GradientBoostingClassifier
model=GradientBoostingClassifier()
model.fit(X_train,y_train)
y_pred=model.predict(X_test)
cm=confusion_matrix(y_pred,y_test)
sns.heatmap(cm,annot=True)
print(classification_report(y_test,y_pred))
```
