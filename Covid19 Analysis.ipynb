```python
import pandas as pd 
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("/Users/anirudhsharma/Downloads/covid19_Confirmed_dataset.csv")
```

```python
df.head(10)
```

```python
df.shape
```

```python
df.drop(["Lat","Long"],axis=1,inplace=True)
```

```python
df
```

```python
DF=df.groupby("Country/Region").sum()
```

```python
DF.head(10)
```

```python
DF.shape
```

```python
DF.loc["India"].plot()
DF.loc["China"].plot()
DF.loc["Italy"].plot()
plt.legend()
```

```python
DF.loc["India"].diff().plot()
```

```python
countries=DF.index
max_infection_rates=[]
for i in countries:
    max_infection_rates.append(DF.loc[i].diff().max())
```

```python
DF["max_infection_rates"]=max_infection_rates
```

```python
infection_rates_data=pd.DataFrame(DF["max_infection_rates"])
```

```python
infection_rates_data
```
